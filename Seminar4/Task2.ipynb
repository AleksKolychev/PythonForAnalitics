```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('kc_house_data.csv',encoding='cp1251')
```

```python
df.head()
```

2.1 Изучите распределение домов от наличия вида на набережную

```python
data_wf = df['waterfront'].value_counts()
data_wf
```

```python
plt.figure()
plt.hist(df['waterfront'], bins = 2)
plt.ylabel('Количество домов')
plt.title('Наличие вида на набережную');
```

```python
plt.title('Наличие вида на набережную')
sns.barplot(x=data_wf.index, y=data_wf.values);
```

2.2 Изучите распределение этажей домов

```python
data_floor = df['floors'].value_counts()
data_floor
```

```python
plt.figure()
plt.hist(df['floors'], bins= 10)
plt.title('Распределение этажей')
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов')
```

2.3 Изучите распределение состояния домов

```python
data_condition = df['condition'].value_counts()
data_condition
```

```python
plt.title('Распределение состояния домов')
sns.barplot(x=data_condition.index, y=data_condition.values);
```
