#### 4.3 Работа с датасетом "house price"

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('kc_house_data.csv',encoding='cp1251')
```

```python
df.head()
```

4.1 Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
    - Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure()
sns.heatmap(corr_matrix, annot=True, linewidths=0.5, cmap='coolwarm');
```

Строим матрицу корреляций и видим, что больше всего на стоимость недвижимости влияет жилая площадь (sqft_living) и (grade). Также цена заметно коррелирует с bathrooms и sqft_ above. Это можно увидеть на графиках взаимного распределения ниже. 

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_above'], kind='reg');
```

```python
sns.jointplot(x=df['price'], y=df['bathrooms'], kind='reg');
```

Как ни странно, на стоимость недвижимости почти никак не влияет ее состояние.
