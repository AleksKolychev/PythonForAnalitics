1.1 Скачать данные по ссылке kc-house-data и laptop_price

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('kc_house_data.csv',encoding='cp1251')
```

```python
df.head()
```

1.2 Изучите стоимости недвижимости  

```python
plt.figure()
plt.hist(df['price'], bins= 50)
plt.xlabel('Цена')
plt.ylabel('Количество домов')
plt.title('Стоимость недвижимости');
```

1.3 Изучите распределение квадратуры жилой площади  

```python
plt.figure()
plt.hist(df['sqft_living'], bins= 50)
plt.xlabel('Квадратные метры')
plt.ylabel('Количество домов')
plt.title('Квадратура жилой площади');
```

1.4 Изучите распределение года постройки 

```python
plt.figure()
plt.hist(df['yr_built'], bins= 50)
plt.xlabel('Годы')
plt.ylabel('Количество домов')
plt.title('Год постройки');
```
